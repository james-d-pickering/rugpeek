```python
def BTK_analytical_model(time, S_conc, I_conc, v0, K_m, K_i, k_inact):
    A = 1 + (S_conc/K_m)
    B = I_conc + (K_i/A)
    k_obs = (k_inact * I_conc) / B
    P = (v0/k_obs)*(1 - np.exp(-k_obs*time))
    return P

def BTK_analytical_model_global(time, i, params):
    S_conc = params['S_conc_'+str(i)]
    I_conc = params['I_conc_'+str(i)]
    v0 = params['v0_'+str(i)]
    K_m = params['K_m_'+str(i)]
    K_i = params['K_i_'+str(i)]
    k_inact = params['k_inact_'+str(i)]
    
    output = BTK_analytical_model(time, S_conc, I_conc, v0, K_m, K_i, k_inact)
    return output


def objective(params, x, data):
    ndata, _ = data.shape
    resid = 0.0*data[:]

    for i in range(ndata):
        resid[i, :] = data[i, :] - BTK_analytical_model_global(x, i, params)
    return resid.flatten()

def normalise(x):
    return x/np.max(x)

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx, array[idx]
```

```python
def ODE_system_generic(time, conc, params, i, S_concs, I_concs):
    #JDP define the system of differential equations to solve
    
    #JDP define initial values
    E, ES, EI, EIstar, P = conc

    #JDP define elements of the rate constant matrix
    k1 = params['k1_'+str(i)].value
    k2 = params['k2_'+str(i)].value
    k1p = params['k1p_'+str(i)].value
    k2p = params['k2p_'+str(i)].value
    kcat = params['kcat_'+str(i)].value
    kinact = params['kinact_'+str(i)].value

    #JDP define relationships between rates
    dE = (-k1*S_concs[i] - k2*I_concs[i])*E + (k1p + kcat)*ES + k2p*EI
    dES = k1*S_concs[i]*E + (-k1p-kcat)*ES 
    dEI = k2*I_concs[i] + (-k2p-kinact)*EI
    dEIstar = kinact*EI
    dP = kcat*ES

    odes = np.array([dE, dES, dEI, dEIstar, dP])
    return odes

def ode_solution_global(time, params, i, system, S_concs, I_concs):
    concs_init = params['E0_'+str(i)].value, params['ES0_'+str(i)].value, params['EI0_'+str(i)].value, params['EIstar0_'+str(i)].value, params['P0_'+str(i)].value
    #print(concs_init, time[0], time[-1], params, i, S_concs, I_concs)
    solution = integrate.solve_ivp(system, (time[0], time[-1]), concs_init, args=(params, i, S_concs, I_concs), method='LSODA', t_eval=time)
    
    return solution

def calc_fluorescence_signal(time, params, i, system, S_concs, I_concs):

    solution = ode_solution_global(time, params, i, system, S_concs, I_concs)

    # extract the concentration you need for the fluorescence (last element)
    conc_E = solution.y[0]
    conc_ES = solution.y[1]
    conc_EI = solution.y[2]
    conc_EIstar = solution.y[3]
    conc_P = solution.y[-1]
    fluor_intensity = (params['N_'+str(i)].value)*conc_P + params['I0_'+str(i)].value
    #fluor_intensity = conc_P
    return [fluor_intensity, conc_E, conc_ES, conc_EI, conc_EIstar, conc_P]

def calc_residual(params, time, data, system, S_concs, I_concs):
    ndata, _ = data.shape
    resid = 0.0*data[:]

    for i in range(ndata):
        resid[i,:] = calc_fluorescence_signal(time, params, i, system, S_concs, I_concs)[0] - data[i, :]

    return resid.flatten()

def global_fit(timebases, signals, init_concs, guesses, mins, maxes, system, S_concs, I_concs, fit_algorithm="least_squares"):
    #JDP this function is the main one you call to run a global fit

    #JDP below is defining all the parameters - every trace in "signals" has a set of parameters
    #JDP min and max values are set according to what was passed in mins and maxes
    params = lf.Parameters()
    for idx, _ in enumerate(signals):
        params.add('E0_'+str(idx), value=init_concs[0], vary=False)
        params.add('ES0_'+str(idx), value=init_concs[1], vary=False)
        params.add('EI0_'+str(idx), value=init_concs[2], vary=False)
        params.add('EIstar0_'+str(idx), value=init_concs[3], vary=False)
        params.add('P0_'+str(idx), value=init_concs[4], vary=False)
        
        params.add('k1_'+str(idx), value=guesses[0], min=mins[0], max=maxes[0], vary=True)
        params.add('k2_'+str(idx), value=guesses[2], min=mins[2], max=maxes[2], vary=True)
        params.add('k1p_'+str(idx), value=guesses[1], min=mins[1], max=maxes[1], vary=True)
        params.add('k2p_'+str(idx), value=guesses[3], min=mins[3], max=maxes[3], vary=True)
        params.add('kcat_'+str(idx), value=guesses[4], min=mins[4], max=maxes[4], vary=True)
        params.add('kinact_'+str(idx), value=guesses[5], min=mins[5], max=maxes[5], vary=True)
        
        params.add('I0_'+str(idx), value=0, min=-np.inf, max=np.inf, vary=True)
        params.add('N_'+str(idx), value=1, min=0, max=50, vary=True)

    #JDP below is fixing the rate constants to be the same for all signals, which is the
    #JDP essence of the global fit idea (multiple datasets fit to the same parameters)
    for idx, _ in enumerate(signals):
        if idx == 0: 
            continue
        else:
            params['k1_'+str(idx)].expr = 'k1_0'
            params['k2_'+str(idx)].expr = 'k2_0'
            params['k1p_'+str(idx)].expr = 'k1p_0'
            params['k2p_'+str(idx)].expr = 'k2p_0'
            params['kcat_'+str(idx)].expr = 'kcat_0'
            params['kinact_'+str(idx)].expr = 'kinact_0'
            
    print("Starting fit using", fit_algorithm, "algorithm...")
    #print("Guess for kf1 is", guesses[0], "nM^-1 s^-1. Bounds: (", mins[0], ",", maxes[0],")")
    #print("Guess for kr1 is", guesses[1], "s^-1. Bounds: (", mins[1], ",",maxes[1],")")
    #print("Guess for kf2 is", guesses[2], "nM^-1 s^-1. Bounds: (", mins[2], ",", maxes[2],")")
    #print("Guess for kr2 is", guesses[3], "s^-1. Bounds: (", mins[3],",", maxes[3],")")
    
    #JDP LMFIT library does all the heavy lifting, minimizing a residual using an algorthim defined by user
    result = lf.minimize(calc_residual, params, args=(timebases, signals, system, S_concs, I_concs),
                    method=fit_algorithm, iter_cb = fit_progress)
    print("\n Fit Complete")

    return result

def fit_progress(params, iteration, resid, *args, **kws):
    #JDP just a function to call during the fit (on each iteration) to monitor progress
    resid_sum = np.sum(resid)
    print("Iteration number "+str(iteration)+" current residual is "+str(resid_sum), end='\r')
    return

def normalise(x):
    return x/np.max(x)

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx, array[idx]
```

```python
### Load in data

data_20 = pd.read_excel('./data_from_cot/compound20_noncovalent_setzero_blank.xlsx', header=[1,2])
data_21 = pd.read_excel('./data_from_cot/compound21_covalent_setzero_blank.xlsx', header=[1,2])
erlotinib = pd.read_excel('./data_from_cot/erlotinib_noncovalent_blank.xlsx', header=[1,2])
osimertinib = pd.read_excel('./data_from_cot/osimertinib_blank.xlsx', header=[1,2])

data_files = [data_20, data_21, erlotinib, osimertinib]
data_names = ['compound 20', 'compound 21', 'erlotinib', 'osimertinib']
```

```python
%matplotlib qt5
plt.close('all')
for index, data in enumerate(data_files):
    print(f"Running analysis for inhibitor: {data_names[index]}")
    I_concs = [500 ,250, 125] #62.5]#31.2]# 15.6]
    conc_headers = [str(x)+" nM" for x in I_concs]
    lowlim = 0
    uplim = 100
    upidx, _ = find_nearest(data['Time (mins)'], uplim)
    lowidx, _ = find_nearest(data['Time (mins)'], lowlim)

    time_mins = data['Time (mins)'][lowidx:upidx]

    signals = []

    for i, conc in enumerate(conc_headers):
        keys = data[conc].keys()
        nkeys = len(keys)
        for j, run in enumerate(keys):
            if j == 0:
                signal = np.array(data[conc][run])
            else:
                signal = signal + np.array(data[conc][run])

        signal = signal[lowidx:upidx]/nkeys

        signals.append(signal)
            

#signals = normalise(np.asarray(signals))
    signals = np.asarray(signals)
    time_mins = np.asarray(time_mins)[:,0]*60 #convert to seconds
    I_concs = np.asarray(I_concs) #nM

    S_concs = np.ones_like(I_concs)*10.0*1E3

    params = lf.Parameters()
    for idx, _ in enumerate(I_concs):
        if idx == 0:
            params.add('S_conc_'+str(idx), value=S_concs[idx], vary=False)
            params.add('I_conc_'+str(idx), value=I_concs[idx], vary=False)
            params.add('v0_'+str(idx), value=1, min=1, max=200)
            params.add('K_m_'+str(idx), value=1E6, min=1E3, max=1E8) 
            params.add('K_i_'+str(idx), value=1E6, min=1E1, max=1E8) # in nM
            params.add('k_inact_'+str(idx), value=1E-2, min=1E-8, max=1E2)
        else:
            params.add('S_conc_'+str(idx), value=S_concs[idx], vary=False)
            params.add('I_conc_'+str(idx), value=I_concs[idx], vary=False)
            params.add('v0_'+str(idx), value=1, min=1, max=200)
            params.add('K_m_'+str(idx), expr='K_m_0')
            params.add('K_i_'+str(idx), expr='K_i_0')
            params.add('k_inact_'+str(idx), expr='k_inact_0')

    mini = lf.Minimizer(objective, params, fcn_args=(time_mins, signals), nan_policy='raise')
    #result = mini.minimize(objective, params, args=(time_mins, signals), method='least_squares')
    result = mini.minimize(method='least_squares')

#plt.figure()
#plt.plot(time_mins, signals[0])
#plt.plot(time_mins, result.best_fit)
#print(result.fit_report())
    plt.figure()
    ax = plt.gca()
    for i in range(len(I_concs)):

        y_fit = BTK_analytical_model_global(time_mins, i, result.params)
        plt.plot(time_mins, signals[i], '-', color='C'+str(i))
        plt.plot(time_mins, y_fit, '-', color='C'+str(i), label=conc_headers[i])
    plt.text(4500, ax.get_ylim()[0]+35000, f'$K_i$: {result.params['K_i_0'].value*1E-9:.3E} M')
    plt.text(4500, ax.get_ylim()[0]+20000, f'$k_{{inact}}$: {result.params['k_inact_0'].value:.3E} s^-1')
    ratio = result.params['k_inact_0'].value / (result.params['K_i_0'].value*1E-9)
    plt.text(4500, ax.get_ylim()[0]+5000, f'Ratio: {ratio:.3E} M^-1 s^-1')
    plt.title(data_names[index])
    plt.xlabel('Time [mins]')
    plt.ylabel('Fluorescence')
    plt.legend()
    lf.report_fit(result.params)
   # ci = lf.conf_interval(mini, result)
    #lf.printfuncs.report_ci(ci)

#ci = lf.conf_interval(mini, result)
```
